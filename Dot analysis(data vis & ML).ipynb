```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('adv.csv')
```

```python
# Extract the features and target variable from the DataFrame
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Create a Linear Regression object
regressor = LinearRegression()

# Fit the model on the training data
regressor.fit(X_train, y_train)
```

```python
# Predict the target variable values for the training set
y_train_predicted = regressor.predict(X_train)

# Calculate the residual values
residuals = y_train - y_train_predicted

# Plot the histogram of the residuals
plt.hist(residuals, bins=50)
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
# Predict the target variable values for the testing set
y_test_predicted = regressor.predict(X_test)

# Calculate the RMSE score
rmse = np.sqrt(mean_squared_error(y_test, y_test_predicted))
print("RMSE score:", rmse)

# Calculate the R-squared score
r2 = r2_score(y_test, y_test_predicted)
print("R-squared score:", r2)

# Plot a scatter plot of the actual vs predicted values
plt.scatter(y_test, y_test_predicted)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.show()
```
